# **Úloha 9 – Integrace funkce jedné proměnné**

```python
### Integrace funkce jedné proměnné

import numpy as np
import matplotlib.pyplot as plt

# Polynomiální funkce
def f0(x): 
    return -x**2 + 4

# Logaritmická funkce
def f1(x): 
    return np.log(x + 2)

# Harmonická funkce
def f2(x): 
    return np.cos(np.pi*x) + 1

# Interval
a = -1
b = 2

# Vizualizace funkcí na intervalu (-2, 2)
x = np.linspace(a, b, 50)

plt.plot(x, f0(x), label="f0(x)")
plt.plot(x, f1(x), label="f1(x)")
plt.plot(x, f2(x), label="f2(x)")

plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.show()
```

Analytické řešení:
$\def\doubleunderline#1{\underline{\underline{#1}}}$

$$\int_{-1}^{2}{4-x^2 \, dx}=\left[4x-\frac{x^3}{3}\right]_{-1}^{2}=8-\frac{8}{3}+4-\frac{1}{3}=\doubleunderline{9}$$

$$\int_{-1}^{2}{\ln{(x+2)} \, dx}=\int_{1}^{4}{\ln{(t)} \, dt}=\left[t\ln{(t)}-t\right]_{1}^{4}=4\ln{(4)}-4-(-1)=\doubleunderline{4\ln{(4)}-3}$$

$$\int_{-1}^{2}{\cos{(\pi x)+1} \, dx}=\frac{1}{\pi}\int_{-\pi}^{2\pi}{\cos{(t)}+1 \, dt}=\frac{1}{\pi}\left[\sin{(t)}+t\right]_{-\pi}^{2\pi}=2-(-1)=\doubleunderline{3}$$

```python
# Integrály daných funkcí nyní spočítáme pomocí tří různých metod (trapezoid, Simpson, Romberg)

from scipy import integrate

for i, (f, res) in enumerate([(f0, 9), (f1, 4*np.log(4) - 3), (f2, 3)]):
    print(f"f{i} | trapezoid: {integrate.trapezoid(f(x), x=x)} | chyba: {integrate.trapezoid(f(x), x=x) - res}")
    print(f"f{i} | Simpson: {integrate.simpson(f(x), x=x)} | chyba: {integrate.simpson(f(x), x=x) - res}")
    print(f"f{i} | Romberg: {integrate.romberg(f, a, b)} | chyba: {integrate.romberg(f, a, b) - res}\n\n")

print("Nejpřesnější je u všech funkcí Rombergova metoda.")
print("Nejméně přesná je u všech funkcí lichoběžníková metoda.")
```
